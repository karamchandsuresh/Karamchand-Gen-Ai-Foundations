```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

Loading the Dataset


```python
df = pd.read_csv("Reviews.csv", nrows=50000)

print(df.shape)
df.head()
```

Find Reviews Containing "light"

```python
light_reviews = df[
    df["Text"].str.contains(r"\blight\b", case=False, na=False)
]

print("Total Reviews:", len(light_reviews))
```

Display Some Reviews

```python
for i, review in enumerate(light_reviews["Text"].head(30)):
    print(f"\nReview {i+1}")
    print(review)
```

Create Sentence List

```python
sentences = [
    "This product is very light and easy on the stomach.",
    "Too much light entered the package.",
    "I enjoy this light and healthy snack."
]
```

Count Vectorization

```python
vectorizer = CountVectorizer()

X = vectorizer.fit_transform(sentences)

dense_matrix = X.toarray()

print(dense_matrix)
```

Extract Vectors

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print(vec_1)
print(vec_2)
print(vec_3)
```

Dot Product Similarity

```python
sim_1_2 = np.dot(vec_1, vec_2)

sim_1_3 = np.dot(vec_1, vec_3)

print("Sentence 1 vs Sentence 2 =", sim_1_2)
print("Sentence 1 vs Sentence 3 =", sim_1_3)
```
